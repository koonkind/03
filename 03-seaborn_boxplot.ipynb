### Import Dataset (mpg.csv)

```python
url='https://raw.githubusercontent.com/worachaiSM/tutorial/main/mpg.csv'
df=pd.read_csv(url)
df.head(5)
```

```python
df['gear']=df.trans.str[0].map({'a':'auto','m':'manual'}).astype("category")
df['drv']=df.drv.map({'f':'front','r':'rear','4':'4-wheel'}).astype("category")
df.head()
```

```python
sns.boxplot(data=df[['cty','hwy']])
```

```python
sns.boxplot(data=df[['displ','cyl']])
```

```python
sns.boxplot(data=df,x='drv',y='cty');
```

```python
sns.boxplot(data=df,y='drv',x='cty',order=['front','rear','4-wheel']);
```

```python
sns.boxplot(data=df,y='drv',x='cty',order=['front','rear','4-wheel'])
plt.ylabel('')
plt.xlabel('city wiles per gallon',weight='bold',size=14)
plt.title('mile per gallon by driving system',weight='bold',size=16);
```

```python
sns.boxplot(data=df,y='drv',x='cty',
            hue='year');
```

```python
sns.boxplot(data=df,y='drv',x='hwy',
            hue='year',
order=['front','4-wheel','rear'])
plt.ylabel('')
plt.xlabel('hight way wiles per gallon',weight='bold',size=20)
plt.title('mile per gallon by driving system',weight='bold',size=19);
```

```python
sns.catplot(y='drv',x='cty',data=df,
           kind='box',
           row='year',
           col='gear',order=['front','4-wheel','rear']);
```
